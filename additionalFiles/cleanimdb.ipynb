```python
# Load Imports
import pandas as pd
```

```python
# Load data
df = pd.read_csv('./imdb_top_1000.csv', low_memory=False)
```

```python
df.head()
```

```python
df['Runtime'].unique()
```

```python
df['Runtime'] = df['Runtime'].str.extract('(\d+)')
df['Runtime'].unique()
```

```python
# Set Genre Column to just the first element
genre_lists = df['Genre'].str.split(',')

my_list = []

for list in genre_lists:
  my_list.append(list[0])

df['Genre'] = my_list
df.head()
```

```python
# check star columns
print(sum(df['Star1'].isnull()))
print(sum(df['Star2'].isnull()))
print(sum(df['Star3'].isnull()))
print(sum(df['Star4'].isnull()))
```

```python
# check director column
print(sum(df['Director'].isnull()))
```

```python
df['Meta_score'] = df['Meta_score'].astype('Int64')
```

```python
df['Meta_score'] = df['Meta_score'].fillna(0)
```

```python
df['Genre'].unique()
```

```python
# Export the file
df.to_csv('clean_imdb.csv', encoding='utf-8', index=False)
```
